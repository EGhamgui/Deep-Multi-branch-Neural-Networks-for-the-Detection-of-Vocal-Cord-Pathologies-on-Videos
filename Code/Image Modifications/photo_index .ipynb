## Import Libraries

```python
from PIL import Image
import os
```

```python
for V in ['out_train_1' , 'out_test_1' , 'out_val_1']: # Choose between train , test and validation data 
    Path = 'C:/Users/USER/Desktop/stage ing/stage ing partie 2/New_1/Masks_cycles/' + V +'/' # The path 
    Folders = os.listdir(Path) # Read the folders 
    for F in Folders:
        if 'edge' in F:
            PathIn = Path + F  + '/'
            Folders1 = os.listdir(PathIn)
            for F1 in Folders1 :
                PathIn1=PathIn+F1
                Photos = os.listdir(PathIn1) # Read all the images from the path : PathIn
                Photos.sort() # Sort image index
                PathOut = PathIn1 +'_aux/' # The new path 
                if F1+'_aux/' not in os.listdir(PathIn): # Create the folder 
                    os.mkdir(PathOut)
                i = 1
                for p in Photos:
                    Image.open(PathIn1+'/'+p).save(PathOut + '{0:03}'.format(i) +'.png') # Save the image with the new index
                    i += 1
```
