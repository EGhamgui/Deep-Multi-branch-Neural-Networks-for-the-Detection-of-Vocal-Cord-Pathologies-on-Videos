## Import Libraries

```python
import pandas as pd
from PIL import Image
import os
import numpy as np
```

```python
df1 = pd.read_excel("C:/Users/USER/Desktop/stage ing/Stage ing partie 3/INFO_subjects_out.xlsx",sep=",")
df1 = df1.set_index(['ID_P','ID_Cy'])
df1.head()
```

```python
i= 0 
k= 0
for V in ['train' , 'val' , 'test']: # Choose between train , test and validation data 
    Path = 'C:/Users/USER/Desktop/stage ing/code/article 2020/dataset/' + V +'/' # The path 
    Path_out = 'C:/Users/USER/Desktop/stage ing/Stage ing partie 5/data'
    
    if V == 'val' :
        ch = 'validation'
    else :
        ch = V

    while (df1.iloc[i]['dataset'] == ch ): 
        if df1.iloc[i]['state'] == 1 : 
            k += 1 
            Cy_f = df1.iloc[i]['Cy_from']
            Cy_t = df1.iloc[i]['Cy_to']
            l=0
            
            F1= Path_out+'_image/C_' + str(k) + '/'
            F2= Path_out+'_mask/C_' + str(k) + '/'
            F3= Path_out+'_mask_binary/C_' + str(k) + '/'
            
            if F1  not in os.listdir(Path_out+'_image/'): # Create the folder 
                os.mkdir(F1)
            if F2  not in os.listdir(Path_out+'_mask/'): # Create the folder 
                os.mkdir(F2)
            if F3  not in os.listdir(Path_out+'_mask_binary/'): # Create the folder 
                os.mkdir(F3)

            for j in range (Cy_f,Cy_t+1):
                if  (j != 7990) and (j!=8067) :
                    Image.open(Path + '{0:05}'.format(j)+'_rgb.png').save(F1 + '{0:05}'.format(l)+'.png') # Save the image with the new index
                    im = Image.open(Path + '{0:05}'.format(j)+'_mask.png')
                    im.save(F2 + '{0:05}'.format(l)+'.png') 
                    width, height = im.size
                    im1 = np.zeros((width, height))
                    
                    for x in range(width):
                        for y in range(height):
                            current_color = im.getpixel( (x,y) )
                            if current_color!= 1:
                                im.putpixel( (x,y), 0)
                            else : 
                                im.putpixel( (x,y), 255)

                    im.save(F3 + '{0:05}'.format(l)+'.png') 
                    l+=1
        i+=1
    print(i)
```
